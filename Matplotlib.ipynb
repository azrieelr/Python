```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,10,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('String Title Here')
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'g--')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Row One')

plt.subplot(1,3,3)
plt.plot(y,x,'r*-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Row Two')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.9,0.9])

axes.plot(x,y,'r--')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('COB')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.5,0.4,0.3])

axes1.plot(x,y,'r--')
axes1.set_title('LARGER PLOT')
axes2.plot(x,y,'b.-')
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 3)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_title('AXES')
       
plt.tight_layout()
```

```python
fig,ac = plt.subplots(nrows=2,ncols=1,figsize=(8,3))

ac[0].plot(x,y)

ac[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.5,0.5])
ax.plot(x,x**10,'b.-',label = 'X Label')
ax.plot(x,x**5,'r--',label = 'Y Label')
ax.set_title('FIGURE')
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.legend(loc=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y+30,color='pink',linewidth=1,linestyle='--',marker='o',markersize=10,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue')
ax.plot(x,y+10,color='#FF8C00',linewidth=1,linestyle=':',marker='1',markersize=15
       ,markerfacecolor='red',markeredgewidth=3,markeredgecolor='black') 
ax.plot(x,y+20,color='purple',linewidth=1,linestyle='-.',marker='s',markersize=10
       ,markerfacecolor='green',markeredgewidth=3,markeredgecolor='orange')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',lw=2,ls='--')
#ax.set_xlim(0,4)
#ax.set_ylim(0,4)
plt.scatter(x,y)
```
